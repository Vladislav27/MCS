```python
import numpy as np
import matplotlib.pyplot as plt

#P = 1, 2, 4, 8, 16
#n=1000000 m=1000

p = np.array([1, 2, 4, 8, 16])
t_1 = np.array([0.132104, 0.132104, 0.132104, 0.132104, 0.132104])
t_p = np.array([0.137534, 0.079328, 0.065809, 0.072451, 0.071762])
s_p = t_1 / t_p
e_p = s_p / p

plt.title("T(P)")
plt.plot(p, t_p)
plt.hlines(0.108301, 0, 16, color='g', label='q_sort')
plt.show()

plt.title("S(P)")
plt.plot(p, s_p)
plt.show()

plt.title("E(P)")
plt.plot(p, e_p)
plt.show()
```
